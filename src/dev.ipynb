```python
load_dotenv()
DATABASE_URL = os.getenv("DATABASE_URL")
```

```python
engine = create_engine(DATABASE_URL)

with engine.connect() as connection:
    query = text('SELECT * FROM t;')
    result = connection.execute(query)

    rows = result.fetchall()


for row in rows:
    print(row)
```
